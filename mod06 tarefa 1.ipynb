#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# %matplotlib notebook

df = pd.read_csv('sinasc_RO_2019.csv')
df
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python

plt.close('all')
plt.rc('figure', figsize=(15, 10))
fig, axes = plt.subplots(2, 3)

sns.countplot(x='APGAR1',ax = axes[0,0], data=df, color = 'b')
sns.countplot(x='APGAR5',ax = axes[0,1], data=df, color = 'r')
sns.countplot(x='SEXO',ax = axes[1,1], data=df, color = 'cyan')
sns.countplot(x='GESTACAO', ax=axes[1, 0], data=df, color='k')
sns.countplot(x='ESTCIVMAE',ax = axes[1,2], data=df, color = 'g',)

axes[1, 0].tick_params(axis='x', rotation=45)
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
sns.barplot(x='APGAR5', y='CONSULTAS', data=df, color = 'b')
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.close('all')
plt.rc('figure', figsize=(15, 10))
fig, axes = plt.subplots(2, 2)

sns.countplot(x='SEXO',ax = axes[0,0], data=df, color = 'b')
sns.countplot(x='GESTACAO',ax = axes[1,0], data=df, color = 'r')
sns.countplot(x='ESTCIVMAE',ax = axes[1,1], data=df, color = 'cyan')
sns.countplot(x='CONSULTAS', ax=axes[0, 1], data=df, color='k')


axes[1, 0].tick_params(axis='x', rotation=40)
```
